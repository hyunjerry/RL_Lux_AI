# Beat the Midterm Champion

This is a notebook uses the Kaggle Lux AI environment to run your agent and the Midterm Champion in a game. Your agent is represented by Player 0. Midterm Champion is represented as Player 1.

To Run, unzip your model from your kaggle submission zip file in the same folder as this notebook. So that your folder structure looks like this:

    |-- UNet_attention >

    |--|-- main.py


Replace the first entry in agent variable in the next code section to point to the main.py of your submission code.

Run for a 100 episodes and show that you win more than 50 of such episodes.

```python
from kaggle_environments import make, evaluate
import numpy as np
from tqdm import tqdm

environment = "lux_ai_2021"
steps = []

# Which agents to run repeatedly.  Same as env.run(agents)
# agents = ["./transformer/main.py" , "./UNet_attention/main.py"]
agents = ["./UNet_attention_64/main.py" , "./UNet_attention/main.py"]
#          ---------------------------- <- replace the underlined section with path to your agent

# How many times to run them.
num_episodes = 100

rewards = [[] for i in range(num_episodes)]
envs = [[] for i in range(num_episodes)]
wins = [[] for i in range(num_episodes)]
for i in tqdm(range(num_episodes), total=num_episodes):
  configuration = {"seed": i, "loglevel": 2, "annotations": True}
  envs[i] = make(environment, configuration, steps, debug=False)
  last_state = envs[i].run(agents)[-1]
  rewards[i] = [state.reward if state.reward is not None else 0 for state in last_state]

wins = np.argmax(rewards, axis=1)
games_won = np.count_nonzero(wins == 0)
print(f'Number of games won by your agent: {games_won}')
print('Pass' if games_won > (0.5 * num_episodes) else 'Fail')
```
